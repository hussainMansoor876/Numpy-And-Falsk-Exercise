<h1 style="color:blue">#Grouping</h1>

# By “group by” we are referring to a process involving one or more of the following steps  Splitting the data into groups based on some criteria Applying a function to each group independently Combining the results into a data structure

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar','foo', 'bar', 'foo', 'foo'],
                  'B' : ['one', 'one', 'two', 'three',
                         'two', 'two', 'one', 'three'],
                  'C' : range(1,9),
                  'D' : range(1,17,2)})
df
```

```python
df.groupby('A').sum()
```

```python
df.groupby(['A','B']).sum()
```

<h1 style="color:blue">#Reshaping</h1>

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz','foo', 'foo', 'qux', 'qux'],
                   ['one', 'two', 'one', 'two','one', 'two', 'one', 'two']]))
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
index
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
df
```

```python
df2 = df[:4]
df2
```

# The stack() method “compresses” a level in the DataFrame’s columns.

```python
stacked = df2.stack()
stacked
```

# With a “stacked” DataFrame or Series (having a MultiIndex as the index), the inverse operation of stack() is unstack(), which by default unstacks the last level:

```python
stacked.unstack()
```

```python
stacked.unstack(0)
```

```python
stacked.unstack(1)
```
